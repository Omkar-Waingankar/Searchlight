```python
stop_words = stopwords.words('english')
congress_stop_words = ['quorum', 'speaker', 'unanimous', 'objection', 'consent', 'clerk', 'senator', 'majority', 'amendment', 'department', 'presiding', 'yield', 'adjourn', 'would', 'madam', 'u']
stop_words += congress_stop_words
lem = WordNetLemmatizer()
def clean_speech(speech):
    tokens = word_tokenize(speech)
    words = [token for token in tokens if token.isalpha() and token.lower() not in stop_words]
    lemmatized_words = [lem.lemmatize(word) for word in words]
    return lemmatized_words
```

# Named Entity Recognition

```python
#find top 10 keywords for speaker
speaker_id = int(allspeakers[allspeakers['last_name'] == 'CORKER']['speaker_id'])
speeches = allspeeches[allspeeches['speaker_id'] == speaker_id]['speech_text']

entity_dictionary = {}
ignore_labels = ['DATE', 'ORDINAL', 'CARDINAL', 'TIME', 'PERCENT', 'MONEY', 'QUANTITY']
for speech in speeches:
    spacy_speech = spacy_nlp(speech)
    for ent in spacy_speech.ents:
        if ent.label_ in ignore_labels:
            continue
        ent_text = ent.text
        if ent_text in entity_dictionary.keys():
            entity_dictionary[ent_text] += 1
        else:
            entity_dictionary[ent_text] = 1
```
