# Yearly Speech Frequency

```python
sessions_id = allspeakers[(allspeakers['last_name'] == 'SESSIONS') & (allspeakers['first_name'] == 'Jeff')].speaker_id
```

```python
sessions = allspeeches[allspeeches['speaker_id'] == int(sessions_id)]
sessions['count'] = [1 for i in range(sessions.shape[0])]
```

```python
sessions
```

```python
year_count = sessions.groupby(['year']).sum()[['count']]
```

```python
list(year_count['count'])
```

```python
fig, ax = plt.subplots()
ax.plot(list(year_count.index), list(year_count['count']))
ax.set(xlabel='time (year)', ylabel='speech count',
       title="Jeff Sessions' Speech Frequency")
ax.grid()
plt.show()
```

```python
def first_name_format(first_name):
    if first_name != '':
        first_name = first_name[0].upper() + first_name[1:].lower()
    return first_name

def plot_yearly_frequency(first_name, last_name):
    last_name = last_name.upper()
    first_name = first_name_format(first_name)
    speaker_id = allspeakers[(allspeakers['last_name'] == last_name) & (allspeakers['first_name'] == first_name)].speaker_id
    speaker = allspeeches[allspeeches['speaker_id'] == int(speaker_id)]
    speaker['count'] = [1 for i in range(speaker.shape[0])]
    year_count = speaker.groupby(['year']).sum()[['count']]
    
    fig, ax = plt.subplots()
    ax.plot(list(year_count.index), list(year_count['count']))
    ax.set(xlabel='time (year)', ylabel='speech count',
       title= first_name + " " + first_name_format(last_name) + "'s Speech Frequency")
    ax.grid()
    plt.show()
```

```python
plot_yearly_frequency('Marco', 'Rubio')
```

```python
plot_yearly_frequency('Bob', 'Corker')
```

```python
plot_yearly_frequency('Cory', 'Booker')
```

```python
plot_yearly_frequency('Nancy', 'Pelosi')
```
